### SQL Tasks

1. download data from https://archive.ics.uci.edu/ml/datasets/car+evaluation
2. create database named 'cardatabase'
3. create table named 'car'
4. dump all data from 'car.data' (downloaded file) into table
5. check if all data has been dumped.
6. group all data with respect to first column (i.e 'cond') and count the occurances of each element in same column
7. filter table where the value is 4 in 'maint' column
8. update the value of 2 with 8 in 'maint' column
9. delete table
10. delete database

```python
import mysql.connector as conn
```

```python
#Creating connection variable
mydb=conn.connect(host="localhost",user='root',password='mysql')
```

```python
#creating cursor variable
cur=mydb.cursor()
```

```python
#viewing existing database
cur.execute("show databases")
cur.fetchall()
```

```python
#creating new database
cur.execute("create database cardatabase")
```

```python
#creating table
cur.execute("create table cardatabase.car  (cond CHAR(10),buying CHAR(10),maint CHAR(10),doors CHAR(10),person CHAR(10),lugboot CHAR(10),safety CHAR(10))")
```

```python
import csv
```

```python
#dumping downloaded data into table
with open ("car.data",'r') as a:
    b=csv.reader(a, delimiter='\n')
    for i in b:
        c=i[0].split(',')
        d="insert into cardatabase.car values ('{}','{}','{}','{}','{}','{}','{}')".format(c[0],c[1],c[2],c[3],c[4],c[5],c[6])
        #print(d)
        cur.execute(d)
mydb.commit()
```

```python
cur.execute("use cardatabase")
```

```python
#groupby on 1st column
cur.execute("select cond,count(cond) from car group by cond")
```

```python
#filtering data with maint=4 
cur.execute("select * from cardatabase.car where maint=4")
```

```python
cur.fetchall()
```

```python
cur.execute("update car set maint=5 where maint='5more'")
```

```python
#updating value 2 with 8
cur.execute("update car set maint=8 where maint=2")
```
